**Pregunta 1:** El fichero cotizacion.csv contiene las cotizaciones de las empresas del IBEX35 con las siguientes columnas: 

* Nombre (nombre de la empresa) 
* Final (precio de la acción al cierre de bolsa) 
* Máximo (precio máximo de la acción durante la jornada) 
* Mínimo (precio mínimo de la acción durante la jornada) 
* Volumen (Volumen al cierre de bolsa)
* Efectivo (capitalización al cierre en miles de euros)

Filtrar en un nuevo dataset aquellos registros cuyo promedio entre el "Máximo" y el "Mínimo" sea mayor al "Final" e imprimir el número de registros.



```python
import pandas as pd
import numpy as np
```

```python
cotizaciones = pd.read_csv("cotizacion.csv", sep=";")
cotizaciones.head()
```

```python
#columnas_a_convertir = ["Final","Máximo","Mínimo","Volumen","Efectivo"]
#cotizaciones["columnas_a_convertir"] = cotizaciones["columnas_a_convertir"].str.replace(',', '.').astype(float)
cotizaciones["Final"] = cotizaciones["Final"].str.replace(',', '.')
cotizaciones["Final"] = pd.to_numeric(cotizaciones["Final"])
cotizaciones["Máximo"] = cotizaciones["Máximo"].str.replace(',', '.').astype(float)
cotizaciones["Mínimo"] = cotizaciones["Mínimo"].str.replace(',', '.').astype(float)
cotizaciones["Volumen"] = cotizaciones["Volumen"].str.replace('.', '')
cotizaciones["Volumen"] = pd.to_numeric(cotizaciones["Volumen"])
cotizaciones["Efectivo"] = cotizaciones["Efectivo"].str.replace(',', '.').astype(float)
cotizaciones.head()
```

```python
tipo3=cotizaciones["Volumen"].dtypes
print(tipo3)
```

```python
cotizaciones["Promedio"]=(cotizaciones["Máximo"]+cotizaciones["Mínimo"])/2
cotizaciones
```

Filtrar en un nuevo dataset aquellos registros cuyo promedio entre el "Máximo" y el "Mínimo" sea mayor al "Final" e imprimir el número de registros.

```python
df_cotizaciones = cotizaciones[cotizaciones["Promedio"]>cotizaciones["Final"]]
df_cotizaciones
```

**Pregunta 2:** Dado el dataset anterior filtrar el "Efectivo" mediante el método between los valores entre 11561.95 y 33000. No incluir los limites.

```python
df_efectivo = df_cotizaciones[df_cotizaciones['Efectivo'].between(11561.95, 33000, inclusive='neither')]
df_efectivo
```

**Pregunta 12:** Dado el dataset titanic:

* Se desea saber que porcentaje de sobrevivientes fueron mujeres(en general).
* Se desea saber que porcentaje de sobrevivientes fueron infantes-niños(0-12 años).


```python
mujeres_supervivientes=(titanic_df[(titanic_df['Survived']==1) & (titanic_df['Sex']=='female')]).shape[0]
porcentaje=(mujeres_supervivientes/len(titanic_df))*100
porcentaje_redondeado = round(porcentaje)
print('El porcentaje de mujeres que sobrevivieron fue del:', porcentaje_redondeado, '%')
```

```python
infantes_niños_supervivientes = (titanic_df[(titanic_df['Survived']==1) & (titanic_df['Age'].between(0,12, inclusive='both'))]).shape[0]
porcentaje_infantes_niños_supervivientes = (infantes_niños_supervivientes/len(titanic_df))*100
porce_redondeado_inf_niños_supervi = round(porcentaje_infantes_niños_supervivientes)
print('El porcentaje de infantes y niños que sobrevivieron fue del:', porce_redondeado_inf_niños_supervi, '%')
```
